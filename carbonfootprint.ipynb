```python
# Carbon Footprint Calculator

def carbon_footprint_calculator():
    print("Carbon Footprint Calculator (Estimates in kg CO₂ per month)")

    # Inputs from user
    electricity_kwh = float(input(" Monthly electricity usage (kWh): "))
    car_km = float(input(" Car travel distance (km): "))
    flights_hours = float(input(" Flight hours: "))
    diet_meat_days = int(input(" Number of meat-heavy meals per week: "))

    # Emission factors (approximate)
    electricity_emission = electricity_kwh * 0.417  # kg CO₂ per kWh
    car_emission = car_km * 0.192  # kg CO₂ per km (average gas-powered car)
    flights_emission = flights_hours * 90  # kg CO₂ per flight hour (average)
    diet_emission = diet_meat_days * 7 * 1.5  # kg CO₂ per day

    # Total
    total_emission = electricity_emission + car_emission + flights_emission + diet_emission

    # Breakdown
    emissions = {
        'Electricity': electricity_emission,
        'Car Travel': car_emission,
        'Flights': flights_emission,
        'Diet': diet_emission,
        'Total': total_emission
    }

    print("\n Estimated Monthly Carbon Footprint (kg CO₂):")
    for key, value in emissions.items():
        print(f"{key}: {value:.2f} kgCO2")

    return emissions

# Run the calculator
emissions_result = carbon_footprint_calculator()
```

```python
import matplotlib.pyplot as plt

# Remove 'Total' from the result for visualization
emissions_to_plot = {k: v for k, v in emissions_result.items() if k != 'Total'}

# Pie chart labels and values
labels = list(emissions_to_plot.keys())
sizes = list(emissions_to_plot.values())
colors = ['#66c2a5','#fc8d62','#8da0cb','#e78ac3']

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title("Carbon Footprint Breakdown (Monthly, kg CO₂)", fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures the pie is a circle.
plt.show()
```
